```python
df = pd.read_csv("data/raw/superstore_raw.csv")
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.dropna(subset=["Sales", "Order Date"])
```

```python
df = df.drop_duplicates()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month
```

```python
df["Profit_Margin"] = df["Profit"] / df["Sales"]
```

```python
df.head()
```

```python
df.to_csv(
    "data/clean/superstore_clean.csv",
    index=False
)
```

```python
!ls data/clean
```

```python
df_curated = df[[
    "Order ID",
    "Order Date",
    "Year",
    "Month",
    "Category",
    "Sub-Category",
    "Region",
    "Sales",
    "Profit",
    "Profit_Margin"
]]
```

```python
df_curated.to_csv(
    "data/curated/superstore_curated.csv",
    index=False
)
```
